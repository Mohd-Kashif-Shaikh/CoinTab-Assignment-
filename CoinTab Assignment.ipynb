```python
def weight_slab(wt):
    if wt<=0.5:
        return 0.5
    elif wt>0.5 and wt<=1:
        return 1
    elif wt>1 and wt<=1.5:
        return 1.5
    elif wt>1.5 and wt<=2:
        return 2
    else:
        return int(wt)+0.5
```

```python
Merge_db['Weight Slab company X']=Merge_db['Total Weight (kg)'].apply(weight_slab)
Merge_db.head()
```

```python
sns.countplot(Merge_db["Weight Slab company X"])
plt.show()
```

# Charges:-

```python
total_charge=[]
for i in range(len(Merge_db_x_pin_courier)):
    fwd_category='fwd_'+Merge_db_x_pin_courier['Zone as per X'][i]
    fwd_fixed=courier_comp_rate['Forward Fixed Charge'][0]
    fwd_additional=courier_comp_rate['Forward Additional Weight Slab Charge'][0]
    rto_category='rto_'+Merge_db_x_pin_courier['Zone as per X'][i]
    rto_fixed=courier_comp_rate['RTO Fixed Charge'][0]
    rto_additional=courier_comp_rate['RTO Additional Weight Slab Charge'][0]
    if Merge_db_x_pin_courier['Type of Shipment'][i]=='Forward charges':
        if Merge_db_x_pin_courier['Weight Slab company X'][i]<=0.5:
            total_charge.append(fwd_fixed)
        else:
            additional_weight=(Merge_db_x_pin_courier['Weight Slab company X'][i]-0.5)/0.5
            total_charge.append(fwd_fixed+additional_weight*fwd_additional)
    if Merge_db_x_pin_courier['Type of Shipment'][i]=='Forward and RTO charges':
        if Merge_db_x_pin_courier['Weight Slab company X'][i]<=0.5:
            total_charge.append(fwd_fixed+rto_fixed)
        else:
            additional_weight=(Merge_db_x_pin_courier['Weight Slab company X'][i]-0.5)/0.5
            total_charge.append(fwd_fixed+additional_weight*(fwd_additional+rto_additional))

Merge_db_x_pin_courier['Expected Charge As Per X (Rs.)']=total_charge
```

```python
result_output['Difference Between Expected Charges and Billed Charges (Rs.)']=result_output['Expected Charge As Per X (Rs.)']-result_output['Charges Billed by Courier Company (Rs.)']
```

```python
count1=result_output[result_output['Difference Between Expected Charges and Billed Charges (Rs.)']==0]['Difference Between Expected Charges and Billed Charges (Rs.)']
total_count1=len(count1)
amount1=count1.sum()
count2=result_output[result_output['Difference Between Expected Charges and Billed Charges (Rs.)']<0]['Difference Between Expected Charges and Billed Charges (Rs.)']
total_count2=len(count2)
amount2=abs(result_output['Difference Between Expected Charges and Billed Charges (Rs.)'].sum())
count3=result_output[result_output['Difference Between Expected Charges and Billed Charges (Rs.)']>0]['Difference Between Expected Charges and Billed Charges (Rs.)']
total_count3=len(count3)
amount3=count3.sum()

dict2={'':['Total orders where X has been correctly charged','Total orders where X has been overcharged','Total orders where X has been undercharged'],
        'Count':[total_count1,total_count2,total_count3],
        'Amount(Rs.)':[amount1,amount2,amount3]}
```

```python
output_data_2=pd.DataFrame(dict2,index=[0,1,2])
output_data_2
```
